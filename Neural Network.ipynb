```python
import torch
# conda activate pytorch

```

```python
def activation (x):
    return 1/(1+torch.exp(-x))
```

```python
# Single layer neural network
torch.manual_seed(7)
feature= torch.randn(1,5)
weight=torch.randn_like(feature)

bias= torch.randn(1,1)
```

```python
y= activation(torch.sum(feature * weight)+bias)
print(y)
```

```python
y= activation((feature*weight).sum()+bias)
print(y)
```

```python
y= activation(torch.mm(feature , weight.view(5,1))+bias)
print(y)
```


# Single Layerd Neural Network

```python
#multi layer neural network
torch.manual_seed(7)
feature= torch.randn((1,3))
n_input=feature.shape[1]
n_hidden=2
n_output=1
w1=torch.randn(n_input,n_hidden)
w2=torch.randn(n_hidden,n_output)

b1= torch.randn((1,n_hidden))
b2= torch.randn(1,n_output)
```

```python
h=activation(torch.mm(feature,w1)+b1)
output=activation(torch.mm(h,w2)+b1)
print(output)
```
